## 7. Group Similar Restaurant Primary Types

#### Infer Specific Restaurant Types from Generic "Restaurant" Labels
The majority of restaurants have a primary type labeled as "restaurant", which lacks sufficient specificity for analysis.
To better capture the nature of each establishment, the types column was leveraged to infer more granular restaurant categories.

```python
df['primaryType'].value_counts().sort_values(ascending=False)
```

```python
primary_types = set(df['primaryType'].unique())
primary_types.discard('restaurant')

types = df['types'].explode().str.replace('_', ' ')
types = types[(types.isin(primary_types)) & (df['primaryType'] == 'restaurant')]
types
```

```python
df['possible_types'] = types.groupby(level=0).first()
df.loc[(df['primaryType']=='restaurant') & df['possible_types'], 'primaryType'] = df['possible_types']

df['primaryType'].value_counts()
```

#### Create a Refined cuisine_group from primaryType
The original primaryType column contains highly granular restaurant categories (e.g. sushi restaurant, pizza restaurant, bar and grill). To reduce fragmentation while preserving meaningful distinctions, similar restaurant types were first grouped into a refined cuisine_group.

This step consolidates closely related business types under a more interpretable category.
For example:
- sushi restaurant and japanese restaurant → japanese restaurant
- pizza restaurant and italian restaurant → italian restaurant
- coffee shop and cafe → beverage shop

Restaurant types that did not belong to any predefined group were retained using their original primaryType to avoid information loss.

```python
american = ['steak house', 'hamburger restaurant', 'diner', 'american restaurant']
italian = ['italian restaurant', 'pizza restaurant']
japanese = ['sushi restaurant', 'japanese restaurant']
middle_eastern = ['middle eastern restaurant', 'lebanese restaurant', 'turkish restaurant']
bar = ['bar','bar and grill', 'night club', 'pub']
beverage = ['coffee shop', 'cafe', 'tea house', 'juice shop']
bakery_dessert = ['bakery', 'ice cream shop']
fast_food = ['fast food restaurant', 'sandwich shop']

df.loc[df['primaryType'].isin(american), 'cuisine_group'] = 'american restaurant'
df.loc[df['primaryType'].isin(italian), 'cuisine_group'] = 'italian restaurant'
df.loc[df['primaryType'].isin(japanese), 'cuisine_group'] = 'japanese restaurant'
df.loc[df['primaryType'].isin(middle_eastern), 'cuisine_group'] = 'middle eastern restaurant'
df.loc[df['primaryType'].isin(bar), 'cuisine_group'] = 'bar'
df.loc[df['primaryType'].isin(beverage), 'cuisine_group'] = 'beverage shop'
df.loc[df['primaryType'].isin(bakery_dessert), 'cuisine_group'] = 'bakery and dessert'
df.loc[df['primaryType'].isin(fast_food), 'cuisine_group'] = 'fast food restaurant'
df['cuisine_group'] = df['cuisine_group'].fillna(df.loc[df['cuisine_group'].isna(), 'primaryType'])
df['cuisine_group'].value_counts()
```

#### Define High-Level main_type Categories
While `cuisine_group` improves interpretability, the number of categories is still relatively large for high-level exploratory analysis. Therefore, a second level of grouping was introduced to classify restaurants into broader `main_type` segments based on cuisine style or business model.

The following high-level categories were defined:
- Western cuisine (e.g. American, Italian, French, Greek)
- Asian cuisine (e.g. Japanese, Chinese, Vietnamese, Indian)
- Beverage and bar (e.g. bars, cafés, bakeries)
- Fast or casual dining (e.g. fast food, buffet, breakfast restaurants)
- Delivery or takeaway
- Other cuisine (categories with relatively small sample sizes)

Restaurants that could not be clearly assigned to any of these groups were labeled as generic restaurant.

This hierarchical grouping strategy balances analytical clarity and flexibility:
- primaryType preserves granular information
- cuisine_group enables cuisine-level comparison
- main_type supports robust, high-level exploratory analysis

```python
western = ['american restaurant', 'italian restaurant', 'french restaurant','greek restaurant', 'mediterranean restaurant']
asian = ['japanese restaurant', 'chinese restaurant', 'vietnamese restaurant', 'indian restaurant', 'thai restaurant',
         'korean restaurant', 'asian restaurant']
beverage_and_bar = ['bar', 'beverage shop','bakery and dessert']
fast_casual = ['fast food restaurant', 'breakfast restaurant', 'buffet restaurant']
delivery_takeaway = ['meal delivery', 'meal takeaway']
other = ['mexican restaurant', 'middle eastern restaurant', 'seafood restaurant', 'vegan restaurant']

# Define the main type of the restaurant according to the cuisine group
df.loc[df['cuisine_group'].isin(western), 'main_type'] = 'western cuisine'
df.loc[df['cuisine_group'].isin(asian), 'main_type'] = 'asian cuisine'
df.loc[df['cuisine_group'].isin(beverage_and_bar), 'main_type'] = 'beverage and bar'
df.loc[df['cuisine_group'].isin(fast_casual), 'main_type'] = 'fast or casual dining'
df.loc[df['cuisine_group'].isin(delivery_takeaway), 'main_type'] = 'delivery or takeaway'
df.loc[df['cuisine_group'].isin(other), 'main_type'] = 'other cuisine'
df['main_type'] = df['main_type'].fillna('generic restaurant')
df['main_type'].value_counts()
```

### Regular Opening Hours Column

```python
opening_hrs_df = df['regularOpeningHours'].apply(lambda x: x.get('periods') if isinstance(x, dict) else None).to_frame()
opening_hrs_df.head()
```

The data in the regularOpeningHours column is a list of dictionaries that contained the open and close time of a period. Thus, the following functions are created in order to calculate the opening hours in each day of a week and to identify if it opens 24 hours.

```python
WEEKDAY_MAP = {0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday'}
WEEK = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
WEEKDAY = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
WEEKEND = ['Sunday', 'Saturday']

def get_weekday(time_dict):
    return WEEKDAY_MAP[time_dict['day']]

def get_timedelta(time_dict):
    return pd.Timedelta(hours=time_dict['hour'], minutes=time_dict['minute'])

# Calculate the opening hours of a day
def calculate_opening_hrs(open_period_dict):
    open_time = get_timedelta(open_period_dict['open'])
    close_time = get_timedelta(open_period_dict['close'])
    duration = close_time - open_time
    
    if duration < pd.Timedelta(0,'d'):
        duration += pd.Timedelta(days=1)

    return duration.total_seconds() / 3600

# Calculate the daily operating hours of a restaurant
def calculate_weekday_operating_hours(raw_open_periods):
    if not raw_open_periods: return None

    result = {'Sunday': 0, 'Monday': 0, 'Tuesday':0, 'Wednesday':0, 'Thursday':0, 'Friday':0, 'Saturday':0}
    for open_period_dict in raw_open_periods:
        if 'open' not in open_period_dict or 'close' not in open_period_dict:
            continue

        weekday = get_weekday(open_period_dict['open'])
        opening_hours = calculate_opening_hrs(open_period_dict)
        result[weekday] += opening_hours
    return result
```

```python
# Check if a restaurant is open 24 hours according to the operation time
def is_open_24_hours(operation_time):
    if not isinstance(operation_time, list):
        return False

    if len(operation_time) != 1:
        return False

    entry = operation_time[0]

    return (
        'open' in entry and
        'close' not in entry and
        entry['open'].get('hour') == 0 and
        entry['open'].get('minute') == 0
    )
```

```python
# Calculate operating_hours
daily_open_hrs_df = opening_hrs_df['regularOpeningHours'].apply(calculate_weekday_operating_hours).apply(pd.Series)
opening_hrs_df = pd.concat([opening_hrs_df, daily_open_hrs_df], axis=1)

# Check if open 24 hours
opening_hrs_df['open_24hrs'] = opening_hrs_df['regularOpeningHours'].apply(is_open_24_hours)
opening_hrs_df.loc[opening_hrs_df['open_24hrs'], WEEK] = 24
opening_hrs_df.head()
```

The opening hours information was extracted and transformed into meaningful features to better represent restaurant availability patterns.

```python
# Weekday & Weekend Open Days
opening_hrs_df['weekday_open_days'] = (opening_hrs_df[WEEKDAY] > 0).sum(axis=1)
opening_hrs_df['weekend_open_days'] = (opening_hrs_df[WEEKEND] > 0).sum(axis=1)

# Total Weekly Opening Hours
opening_hrs_df['total_hrs'] = opening_hrs_df[WEEK].sum(axis=1)

# Avg. Daily, Weekday, Weekend Hours
opening_hrs_df['avg_daily_hrs'] = opening_hrs_df['total_hrs'] / (opening_hrs_df['weekday_open_days'] + opening_hrs_df['weekend_open_days'])
opening_hrs_df['avg_weekday_hrs'] = opening_hrs_df[WEEKDAY].sum(axis=1) / opening_hrs_df['weekday_open_days']
opening_hrs_df['avg_weekend_hrs'] = opening_hrs_df[WEEKEND].sum(axis=1) / opening_hrs_df['weekend_open_days']

opening_hrs_df.head()
```

```python
# Add to the main df
opening_hrs_df = opening_hrs_df[['open_24hrs', 'weekday_open_days','weekend_open_days', 'total_hrs', 'avg_daily_hrs', 'avg_weekday_hrs','avg_weekend_hrs']]
df = pd.concat([df, opening_hrs_df], axis=1)
df.head()
```
